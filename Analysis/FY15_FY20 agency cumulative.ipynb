```python
# import raw data
rawImport = pd.read_excel('input/FY15_FY20.xlsx')
rawImport
```

```python
#remove all rows when "solicitation review" does not equal 1

reviewableSol = rawImport.loc[rawImport['Solicitation Reviewable'] == 1]
reviewableSol
```

```python
# only need these cols: Date, FY, agency, office, compliant dummy, and effective compliance dummy
cleanDf = reviewableSol[['PostedDate','FY','AGENCY','OFFICE','Relevant','Compliant Dummy','Effective Compliance Dummy','OFFICE_Corrected']].reset_index(drop = True)
#cleanDf['AGENCY'][1]
cleanDf
```

```python
#extract the agency we are interested in (agency with the most solicitations, top 15 agencies)
totalSol = cleanDf.groupby(['AGENCY']).sum()['Relevant'].reset_index() #sort by volume
mask1 = totalSol['Relevant']>42
#minium number of solicitations an agency must have to be included in the graph or dataframe

mask1 
#will return True or False, then we will only keep the agency where there greater than 20 solicitations 
```

```python
top15 = totalSol.loc[mask1,:].copy().reset_index(drop = True)
top15
```

```python
sol = sum(top15['Relevant'])
sol #number of relevant solicitations
```

# The dataframe created below(finalDf) will be the starting point for all graphs and charts related to comparing agencies. 
It ONLY include agencies with more than 20 solicitations cumulatively, reviewable solicitations (therefore cannot be used when graphing for product related because subproducts is labled as n/a for reviewable solicitation column. 

```python
#extract from cleanDf only the top15 agencies
#finalDf = []
finalDf = cleanDf.set_index('AGENCY').loc[top15['AGENCY']].reset_index()
finalDf
#finalDf will be used to create tables and graphs
```

# creating a table for cumulative FAR compliance across agencies
#table col headers will be agency, sum of revelvant, sum of Compliant, sum of noncompliant(relevant - compliant), percent compliant. We already found the relevant (top15)

```python
#to find FAR compliant solicitations, 'compliant dummy' = 1
cumulativeFAR = finalDf.groupby(['AGENCY']).sum()['Compliant Dummy'].reset_index()
cumulativeFAR
```

```python
diff = pd.DataFrame(top15['Relevant'] - cumulativeFAR['Compliant Dummy'])
nonCompliant = pd.DataFrame(top15['AGENCY'])
#notCompliant0['AGENCY']=agency
nonCompliant['notCompliant'] = diff
nonCompliant
```

```python
percentCompliant = finalDf.groupby(['AGENCY']).mean()['Compliant Dummy'].reset_index()
percentCompliant['Compliant Dummy'] = pd.Series(["{:.0%}".format(val) 
                                                 for val in percentCompliant['Compliant Dummy']], index = percentCompliant.index)
#convert it to percentage format
percentCompliant
```

```python
#importing effective 

cumulativeEff = finalDf.groupby(['AGENCY']).sum()['Effective Compliance Dummy'].reset_index()
cumulativeEff
```

```python
diffEff = pd.DataFrame(top15['Relevant'] - cumulativeEff['Effective Compliance Dummy'])
nonEffCompliant = pd.DataFrame(top15['AGENCY'])
#notCompliant0['AGENCY']=agency
nonEffCompliant['notEffCompliant'] = diffEff
nonEffCompliant
```

```python
percentEffCompliant = finalDf.groupby(['AGENCY']).mean()['Effective Compliance Dummy'].reset_index()
percentEffCompliant['Effective Compliance Dummy'] = pd.Series(["{:.0%}".format(val) 
                                                 for val in percentEffCompliant['Effective Compliance Dummy']], index = percentEffCompliant.index)
#convert it to percentage format
percentEffCompliant
```

```python
top15['FAR compliant'] = cumulativeFAR['Compliant Dummy']
top15['FAR noncompliant'] = nonCompliant['notCompliant']
top15['percent FAR compliant'] = percentCompliant['Compliant Dummy']

#import effective compliance data into dataframe
top15['Eff compliant'] = cumulativeEff['Effective Compliance Dummy']
top15['Eff noncompliant'] = nonEffCompliant['notEffCompliant']
top15['percent Eff compliant'] = percentEffCompliant['Effective Compliance Dummy']

top15Sorted = top15.sort_values('Relevant',ascending=False).reset_index(drop = True)
top15Sorted
```

```python
#plotting FAR compliant across agency cumulatively 
# import plotly.tools as tls
# tls.set_credentials_file(username='liyangwang', api_key='BtCOcoz2XDxnBRmFaXAX')

x = top15Sorted['AGENCY']
yFAR = top15Sorted['FAR compliant']
y2FAR = top15Sorted['FAR noncompliant']

FAR = pd.DataFrame({'x': x, 'y': yFAR, 'y2':y2FAR})
FAR.head()

#annotation 
textFAR = top15Sorted['percent FAR compliant']

#x = top15Sorted['AGENCY']
yEff = top15Sorted['Eff compliant']
y2Eff = top15Sorted['Eff noncompliant']

Eff = pd.DataFrame({'x': x, 'y': yEff, 'y2':y2Eff})
Eff.head()

#annotation 
textEff = top15Sorted['percent Eff compliant']
```
